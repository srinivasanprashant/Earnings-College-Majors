```python
# Setup the environment by importing the libraries we need and running the necessary Jupyter magic so that plots are displayed inline
```

```python
%matplotlib inline  

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset into a DataFrame and start exploring the data.
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
```

```python
print(recent_grads.head())
```

```python
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

Drop rows with missing values. Matplotlib expects that columns of values we pass in have matching lengths and missing values will cause matplotlib to throw errors.

```python
# Look up the number of rows in recent_grads
raw_data_count = recent_grads.count()
print("Raw Data Count:", max(raw_data_count))
# drop rows containing missing values
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.count()
print("Cleaned Data Count:", max(cleaned_data_count))
```

# Generate scatter plots to explore the following relations:
- Sample_size and Employed
- Sample_size and Median
- Sample_size and Unemployment_rate
- Full_time and Median
- ShareWomen and Unemployment_rate
- Men and Median
- Women and Median

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
# Sample_size and Median
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size')
```

```python
# Sample_size and Unemployment_rate
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Unemployment_rate vs. Sample_size')
```

```python
# Full_time and Median
recent_grads.plot(x="Full_time", y="Median", kind="scatter", title="Full_time and Median")
```

```python
# ShareWomen and Unemployment_rate
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter", title="ShareWomen and Unemployment_rate")
```

```python
# Men and Median
recent_grads.plot(x="Men", y="Median", kind="scatter", title="Men and Median")
```

```python
# Women and Median
recent_grads.plot(x="Women", y="Median", kind="scatter", title="Women and Median")
```

```python
# ShareWomen and Unemployment_rate
recent_grads.plot(x="ShareWomen", y="Median", kind="scatter", title="ShareWomen and Median")
```

Students in more popular majors do not make more money. In fact, it appears from the data that they make less median income vs. majors that are not as popular.

Students that majored in subjects that were majority female make less money. As the share of women tends towards 100%, the median income tends to decrease.
